## Matplotlib example workflow

```python
# A matplotlib workflow

# 0. Import and get matplotlib ready
import matplotlib.pyplot as plt

# 1. Prepare / Get your data
x= [1,2,3,4]
y=[11,22,33,44]

# 2. Setup plot ( instances of the objects)
fig,ax = plt.subplots(figsize=(10,10)) # 

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Sample plot" , 
       xlabel="X-Axis",
       ylabel="Y-Axis")

# 5. Save & Show
fig.savefig("numpy_images/simple-sample-plot.png")

       
```

## Making figures with Numpy arrays

Most of figuring out what kind of plot to use is getting a feel for the data, then see what suits it best.

Matplotlib visualizations are built off NumPy arrays. So in this section we'll build some of the most common types of plots using NumPy arrays.

* line plot (default)
* scatter plot
* bar plot
* histogram
* subplots()

# Line 

```python
# Create an array
x=np.linspace(0,10,100) # if confusing, hower over inside the () and press shift+tab to read documentation
x[:10]
```

```python
# The default kind of plot is a line plot
fig,ax = plt.subplots()
ax.plot(x,x**2)
```

# Scatter 

```python
# We need to create our figure and axis instance when we want a new figure
fig, ax = plt.subplots()
ax.scatter(x,np.exp(x))
```

```python
fig,ax = plt.subplots()
ax.scatter(x,np.cos(x))
```

# Bar Plot
 * Horizontal
 * Vertical

Vertical

```python
# You can make plots from a dictionary, just ensure you pass in the correct values
sanjiv_fav_food_prices = {"Special Shawarma": 200,
                          "Benne Dosa":100,
                           "Sea Food Platter": 750}
fig, ax = plt.subplots()
ax.bar(sanjiv_fav_food_prices.keys(), sanjiv_fav_food_prices.values())
```

Horizontal

```python
fig,ax = plt.subplots()
ax.barh(sanjiv_fav_food_prices.keys(), sanjiv_fav_food_prices.values())
ax.set(title="Sanjivs Favourite Dishes",
       xlabel="Cost in INR")
```
