## Explanation of these shapes and pixel values

432 → height (number of rows, i.e. pixels vertically)

575 → width (number of columns, i.e. pixels horizontally)

4 → channels (the per-pixel values, in your case R, G, B + Alpha/Transparency)


Take a tiny 2×2 RGB image:

[
  [ [255, 0, 0],     [0, 255, 0]   ],
  [ [0, 0, 255],     [255, 255, 0] ]
]


Top left pixel → pure red (255,0,0)

Top right pixel → green (0,255,0)

Bottom left pixel → blue (0,0,255)

Bottom right pixel → yellow (255,255,0)

Shape → (2, 2, 3)

In your case

dog.shape = (432, 575, 4) means:

You’ve got 432 rows of pixels

Each row has 575 pixels across

And each pixel is described by 4 values → [R, G, B, Alpha].
